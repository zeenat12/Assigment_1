## Q4. What is a multiprocessing pool in python? Why is it used?

A multiprocessing pool in Python is a high-level abstraction provided by the multiprocessing module that allows you to easily manage and distribute the execution of tasks across multiple processes. A pool maintains a group of worker processes that can be used to execute tasks in parallel. It abstracts away much of the complexity involved in creating and managing individual processes, making it simpler to achieve parallelism in your code.

The main purpose of using a multiprocessing pool is to efficiently parallelize the execution of a function across a set of input values. Here's how it works:

Creating a Pool: You create a pool of worker processes using the Pool class from the multiprocessing module. You specify the number of worker processes you want in the pool. Each worker process is a separate process that can execute tasks concurrently.

Mapping Tasks: Once the pool is created, you can use its map() or imap() method to apply a given function to a collection of input values. The pool automatically distributes the tasks across the worker processes, ensuring that the tasks are executed in parallel.

Results: The map() method returns a list of results, where each result corresponds to the output of applying the function to each input value. If you use imap(), you get an iterator over the results, which can be useful for processing large amounts of data without loading all the results into memory at once.

## Q5. How can we create a pool of worker processes in python using the multiprocessing module?

```python
import multiprocessing
def square(n):
    return n**2
if __name__ == '__main__':
    with multiprocessing.Pool(processes=5) as pool :
        out =pool.map(square , [3,4,5,6,6,7,87,8,8])
        print(out)
```

## Q6. Write a python program to create 4 processes, each process should print a different number using the multiprocessing module in python.

```python
import multiprocessing

def print_number(number):
    print(f"Process {number}: {number}")

if __name__ == "__main__":
    processes = []
    
    for i in range(1, 5):
        process = multiprocessing.Process(target=print_number, args=(i,))
        processes.append(process)
        process.start()
    
    for process in processes:
        process.join()
```
