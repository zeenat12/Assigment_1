# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

**1 . Scatter plot:**

**In seaborn, scatter plot are used for**

1. **Visualizing Relationships**: Displaying relationships and patterns between two continuous variables.

2. **Correlation Analysis**: Assessing the strength and direction of correlations.

3. **Outlier Detection**: Identifying outliers or unusual data points.

4. **Pattern Recognition**: Spotting trends or clustering in data.

5. **Comparing Data Groups**: Comparing relationships across different categories or groups.

6. **Regression Analysis**: Overlaying regression lines to model relationships.

7. **Anomaly Detection**: Detecting anomalies or deviations from the typical scatter pattern.

8. **Interactive Visualization**: Creating interactive scatter plots for data exploration.

**2. Histogram**

In Seaborn, histograms are used for:

1. **Visualizing Data Distribution**: Displaying the frequency distribution of a single continuous variable.

2. **Identifying Data Patterns**: Assessing data patterns such as symmetry, skewness, and modes.

3. **Understanding Data Spread**: Gauging how data values are spread across a range.

4. **Detecting Outliers**: Identifying potential outliers or extreme values in the data.

5. **Comparing Multiple Distributions**: Comparing the distributions of multiple variables or groups.

**3. relational plots**

In Seaborn, relational plots (e.g., scatter plots, line plots) are used for:

1. **Visualizing Relationships**: Displaying relationships and patterns between two or more variables.

2. **Correlation Analysis**: Assessing the strength and direction of correlations between variables.

3. **Trend Analysis**: Examining trends and changes in data over time or across conditions.

4. **Comparing Multiple Variables**: Comparing relationships or trends across different categories or groups.

5. **Outlier Detection**: Identifying outliers or unusual data points in the context of relationships or trends.

6. **Visualizing Complex Data**: Handling complex datasets with multiple variables and exploring their interactions.

**4. categorical plot**

In Seaborn, `catplot` (categorical plot) is used for:

1. **Visualizing Categorical Data**: Displaying relationships and distributions involving categorical variables.

2. **Comparing Categories**: Comparing data across different categories or groups.

3. **Showing Data Distributions**: Visualizing the distribution of data within categories.

4. **Combining Multiple Plot Types**: Creating a variety of plot types (e.g., bar plots, box plots) to analyze categorical data effectively.

5. **Faceting**: Faceting allows you to create subplots for each category, making it easier to compare and analyze multiple aspects of your data.

**5. joint plot**

In Seaborn, `jointplot` is used for:

1. **Bivariate Analysis**: Exploring the relationship between two numeric variables.

2. **Visualizing Distributions**: Showing the individual distributions of both variables.

3. **Correlation Assessment**: Displaying scatter plots, regression lines, and correlation coefficients.

4. **Hexbin Plots**: Visualizing data density using hexagonal binning.

5. **Kernel Density Estimation**: Estimating and visualizing the data's probability density functions.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.


## Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
```

```python
fmri
```

```python
sns.lineplot( x= "timepoint" , y = "signal" , data = fmri)
plt.show()
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

# Note: pclass, age, and fare are columns in the titanic dataset.

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic
```

```python
sns.boxenplot(x = "pclass" , y = "age" , data = titanic)
plt.show()
```

```python
sns.boxenplot(x = "pclass" , y = "fare" , data = titanic)
plt.show()
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.
