# Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

Pandas DataFrame,is a two-dimensional structure with columns and rows whose size can be changed. You can also think of it as a dictionary of Series objects. 

a Pandas Series is a 1D array of data, but a single-column DataFrame is a 2D table with one column. 

In pandas, a DataFrame is a two-dimensional, tabular data structure that is designed to store and manipulate data in a way that resembles a spreadsheet or SQL table. It consists of rows and columns, where each column can contain data of a different data type. DataFrames are one of the most commonly used data structures in pandas, and they offer powerful data manipulation and analysis capabilities.

Here are some key characteristics of a DataFrame:

1. **Two-Dimensional Structure:** DataFrames are two-dimensional, meaning they have rows and columns. You can think of them as similar to a table in a relational database or an Excel spreadsheet.

2. **Heterogeneous Data Types:** Each column in a DataFrame can have a different data type (e.g., integer, float, string, datetime), which makes them suitable for handling diverse types of data.

3. **Labeled Axes:** DataFrames have labeled rows (index) and columns, which makes it easy to reference and manipulate data based on labels.

4. **Data Alignment:** DataFrames automatically align data based on the index, which simplifies operations involving multiple DataFrames.


In summary, while both Series and DataFrames are fundamental data structures in pandas, the key difference is in their dimensions and intended use. Series are one-dimensional, primarily used for representing a single column of data, while DataFrames are two-dimensional and serve as a container for multiple columns of data. DataFrames are more versatile for handling structured data, while Series are more suitable for one-dimensional data.

# Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

```python
import pandas as pd
df = pd.read_csv('titanic_data.csv')
```

1. **`head()` and `tail()`**:
   - Use `head()` to view the first few rows of a DataFrame.
   - Use `tail()` to view the last few rows of a DataFrame.


```python
# Example:
df.head()
```

```python
df.tail()
```

2. **`shape`**:
   - Returns the dimensions (number of rows and columns) of the DataFrame.

```python
# Example: 
df.shape
```

3. **`info()`**:
   - Provides a summary of the DataFrame, including data types, non-null values, and memory usage.

```python
#Example:
df.info()
```

4. **`describe()`**:
   - Generates summary statistics (count, mean, std, min, max, etc.) for numeric columns.

```python
#Example:
df.describe()
```

5. **`loc[]` and `iloc[]`**:
   - Use `loc[]` to select rows and columns by labels.
   - Use `iloc[]` to select rows and columns by integer positions.

```python
#Example:
df.loc[2:4, ['Survived', 'Pclass']]
```

```python
df.iloc[0:3, 1:3]
```

# Q6. Which of the following is mutable in nature Series, DataFrame, Panel?`

- **Series**: A Pandas Series is mutable, meaning you can modify the elements of a Series after it has been created. You can change the values of individual elements or add new elements to a Series.

- **DataFrame**: Similarly, a Pandas DataFrame is also mutable. You can modify the data in a DataFrame, add or remove columns, and update values within the DataFrame.

- **Panel**: Panel was designed to handle three-dimensional data, but it has been phased out in favor of using MultiIndex DataFrames.
